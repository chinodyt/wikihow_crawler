```python
settings = wc.Settings()
print(settings.settings)
```

```python
output_path = settings.settings['output_path']
query_size = settings.settings['n_query']
for lang in settings.settings['query'] :
    if lang == 'pt' :
        continue
    full_path = os.path.join(output_path, lang)
    if not os.path.isdir(full_path) :
        print(f"Creating '{full_path}'...")
        os.makedirs(full_path)
    crawler = wc.Crawler(lang)
    rs = []
    for q in settings.settings['query'][lang] :
        print(f"Query: {q}")
        rs = crawler.search(q, resultset = rs, n = query_size)
    
    print('Downloading PDFs and saving CSV file...')
    with open(os.path.join(full_path, 'data.csv'), 'w') as f_csv:
        csv_writer = csv.writer(f_csv, delimiter = ';', quotechar = '"', quoting=csv.QUOTE_MINIMAL)
        for r in rs :
            r.save_pdf(full_path)
            text = str(r)
            text = text.replace('\n', '\\n')
            csv_writer.writerow([r.url, r.filename, text])
    

#crawler = wc.Crawler('pt')
#rs = crawler.search('impressora', n = 3)
#rs = crawler.search('instalar impressora', resultset=rs, n = 3)

```
